```python
num_accelerometer_data_points = len(accelerometer_data)
slider_range_and_step = (0, num_accelerometer_data_points, 1)

@interact(start_range=slider_range_and_step, end_range=slider_range_and_step)
def plot_accelerometer_data(start_range=0, end_range=num_accelerometer_data_points):
    fig = pyplot.figure(figsize=(10, 10), dpi=300)
    fig.suptitle("Raw accelerometer data of a sample", fontsize=20)
    subplot = fig.add_subplot(1, 1, 1)
    subplot.tick_params(labelsize=20)
    
    subplot_items = [("x acceleration", 1), ("y acceleration", 2), ("z acceleration", 3)]
    for subplot_label, column_index in subplot_items:
        subplot.plot(
            accelerometer_data[start_range:end_range, 0],
            accelerometer_data[start_range:end_range, column_index],
            
            label=subplot_label, linewidth=2)
    
    subplot.yaxis.grid()
    subplot.set_xlabel("Time passed since notification [s]", fontsize=20)
    pyplot.legend(fontsize=20)
    
    pyplot.show()
    
    # fig.savefig("raw_accelerometer_data_sample.png")
```
