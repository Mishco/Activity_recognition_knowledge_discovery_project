This notebook shows an example of a matrix where each non-zero field represents the similarity of features for different body postures of a single user. The features were computed from touch screen swipes on a smartphone. The similarity was computed as a *t*-test statistic.

```python
%matplotlib inline

import numpy
from matplotlib import pyplot
from matplotlib import figure
```

```python
matrices = []
matrices.append(numpy.matrix("0.      0.0177  0.0553  0.6654  0.0004  0.6464; \
  0.      0.      0.3917  0.0029  0.0234  0.0092; \
  0.      0.      0.      0.0087  0.0051  0.029; \
  0.      0.      0.      0.      0.0001  0.9201; \
  0.      0.      0.      0.      0.      0.0002; \
  0.      0.      0.      0.      0.      0.    "))
```

```python
classes = ["standing", "sitting_straight", "sitting_back_relaxed", "sitting_bent_forward",
           "sitting_with_arms_on_table", "sitting_on_bag"]
fig = pyplot.figure(figsize=(10, 10))
subplot_index = 1
for matrix in matrices:
    subplot = fig.add_subplot(len(matrices), 1, subplot_index)
    if subplot_index == 1:
        subplot.set_xticklabels([''] + classes, rotation=270, fontsize=20)
        subplot.set_yticklabels([''] + classes, fontsize=20)
    else:
        subplot.set_xticklabels([''])
        subplot.set_yticklabels([''])
    matrix_plot = subplot.matshow(matrix)
    colorbar = fig.colorbar(matrix_plot)
    colorbar.set_label("Similarity", fontsize=20)
    colorbar.ax.tick_params(labelsize=20)
    
    subplot_index += 1
```
